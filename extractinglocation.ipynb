```python
import spacy
from data.function import Categories

nlp = spacy.load('en_core_web_sm')

def extract_location(text):
    doc = nlp(text)
    for ent in doc.ents:
        if ent.label_ == "GPE":
            return ent.text
    return None


def noun(text):
    cleantext = text.replace('"','').replace("'","")
    doc = nlp(cleantext)
    group = []
    for i in doc.noun_chunks:
        group.append(i.text)
    return group

def match_stock(noun):
    stock = Categories().get_all_stock()
    matching = []
    for i in noun:
        if i in stock:
            matching.append(i)
    return list(set(matching))
            
def location(noun):
    stock = Categories().get_all_stock()
    possible_location = []
    for i in noun:
        if i in stock:
            continue
        doc = nlp(i)
        for ent in doc.ents:
            if ent.label_ == "GPE" or ent.label_ == "PERSON" or ent.label_ == "ORG":
                possible_location.append(i)
    possible_location = list(set(possible_location))
    return possible_location
```
